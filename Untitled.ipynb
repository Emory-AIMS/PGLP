```python
sample_size = 5000

from scipy.spatial import ConvexHull, convex_hull_plot_2d
points = np.random.rand(30, 2)   # 30 random points in 2-D
hull = ConvexHull(im.sensitivity_hull)

vertice = hull.points
simplices = hull.simplices
segments = []
total_length = 0
for simplex in simplices:
    segment = np.linalg.norm(vertice[simplex[1]] - vertice[simplex[0]])
    segments.append(segment)
    total_length += segment
```

```python
samples = []
for _ in range(sample_size):
    random_length = np.random.rand() * total_length
    temp_total_length = random_length
    for i, segment in enumerate(segments):
        if temp_total_length - segment < 0:
            break
        temp_total_length = temp_total_length - segment

    left_vertex = simplices[i][0]
    right_vertex = simplices[i][1]

    sample = vertice[left_vertex] + (vertice[right_vertex] - vertice[left_vertex]) * temp_total_length / segments[i]
    samples.append(sample)
samples = np.array(samples)

mean_samples = np.average(samples, axis=0)
Y = [np.dot((sample - mean_samples).reshape(-1,1), (sample - mean_samples).reshape(-1,1).T) for sample in samples]
print(np.average(Y, axis=0))
Y = np.power(np.average(Y, axis=0), -1/2)
transformed_vertice = np.dot(Y, vertice.T).T
```
