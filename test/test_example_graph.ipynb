```python
import numpy as np
epsilon = 1
point = np.array([0,1])
```

## Compare utility by the experiment

### WIG
- Handmade example
- Three examples shown in the first step
    
### Mechanisms
- S-LM
- S-PIM
- ELM

### Experiments
- perturbation of one location (0,1) (not trajectry)
- iterate this perturbation

```python
import sys
sys.path.append("../")

import matplotlib.pyplot as plt

def show_utility(mec, n_iter=10000, name="example1"):
    mec.load_from_jbl(name)
    mec.build_distribution(epsilon)
    samples = np.zeros((n_iter,2))
    errors = np.zeros((n_iter,1))
    
    for i in range(n_iter):
        samples[i,:] = mec.perturb(point)
        errors[i,:] = np.linalg.norm(point - samples[i,:])
        
    print("ERROR:", np.average(errors))
    
    plt.hist2d(samples[:,0], samples[:,1], bins=100)
    plt.colorbar()
    plt.xlim(-5,5)
    plt.ylim(-4,6)
    
    
import src.mechanism as MEC
def show_info(name):
    mec = MEC.Mechanism()
    mec.load_from_jbl(name)
    print("locations")
    x = mec.coords[:,0]
    y = mec.coords[:,1]
    plt.scatter(x,y)
    
    for x_, y_, name in zip(x, y, mec.names):
        plt.text(x_, y_, name, fontsize=15)
    
    plt.show()
    print("WIG\n", mec.weight_mat)
```
